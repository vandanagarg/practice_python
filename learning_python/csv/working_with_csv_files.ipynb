**USING REGULAR EXPRESSIONS FOR FILTERING DATA**

```python
import re

df.loc[df['Type 1'].str.contains('Fire|Grass',regex=True)]
```

```python
#ignoring capitalization
df.loc[df['Type 1'].str.contains('fire|grass',flags= re.I, regex=True)]
```

```python
#in case we need to find a few data in which name of pokemon starts from pi and could have any data later on
df.loc[df['Name'].str.contains('^pi[a-z]*',flags= re.I, regex=True)]  # * indicates 0 or more characters 
# ^ indicates start of line/word
# just writing 'pi[a-z]*' indicates pi could be in middle not fixed in starting position
```
